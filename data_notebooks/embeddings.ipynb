```python
from openai import OpenAI
import pandas as pd
import numpy as np

# Modify OpenAI's API key and API base to use vLLM's API server.
openai_api_key = "EMPTY"
openai_api_base = "http://localhost:8001/v1"

client = OpenAI(
    # defaults to os.environ.get("O|PENAI_API_KEY")
    api_key=openai_api_key,
    base_url=openai_api_base,
)
#client.models.list()
model = 'Qwen/Qwen3-Embedding-0.6B'
```

```python
summaries_new = pd.read_csv('../data/summaries_new.csv')
summaries_list = summaries_new.summary.to_list()
```

```python
print(summaries_list[150])
```

```python
responses = client.embeddings.create(
    input=summaries_list,
    model=model,
)
```

```python
embeddings = list(map(lambda x: x.embedding, responses.data))
summaries_new['embeddings']=embeddings
summaries_new.to_csv('../data/embedded_table.csv')
!du -h ../data/embedded_table.csv
summaries_new
```
